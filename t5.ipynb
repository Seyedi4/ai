```python
import math as m
data=[(0,-1),(0.5,1),(1,-1)]
alpha=0.1
b=[2.14,1.29,-0.58]
w=[-34.4,-2.52,-1.30,2.28]
def noron(w,b,x):
        return w*x+b
def active(z):
    a=m.log(1+ m.exp(z))
    return a
def norons(w1,w2,b,a1,a2):
    return w1*a1+w2*a2+b

def fb(y,a,z):
    return -2*(y-a)*m.exp(z)*1/(1+m.exp(z))
def fw(y,a,z,x):
    return -2*(y-a)*m.exp(z)*1/(1+m.exp(z))*x

for x,y in data:
    for i in range(10):
        z1 = noron(w[0], b[0], x)
        a1 = active(z1)
        print(a1)
        print(" /")
        z2 = noron(w[1], b[1], x)
        a2 = active(z2)
        print(a2)
        print(" /")
        z3 = norons(w[2], w[3], b[2], a1, a2)
        a3=active(z3)
        print(a3)
        print("------------")
        b[2]=b[2]-alpha*fb(y,a3,z3)
        w[2]=w[2]-alpha*fw(y,a3,z3,a1)
        w[3]=w[3]-alpha*fw(y,a3,z3,a2)
        
```
